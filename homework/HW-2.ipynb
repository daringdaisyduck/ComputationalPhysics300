```python
#Create a program to make an image of the Mandelbrot set by performing the iteration for all 
#values of c = x+iy

#import the libraries needed
#import numpy for functions
#import pyplot to plot
#%matplotlib inline allows graphs to be included in my notebook
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Declare variable to set the number of iterations for the mandelbrot function.
#Create numpy arrays using linspace for x and y axes bounded between -2 and 2.
#Complex_Values is an array that will store the values of the set of complex values calculated.
#Mandelbrot_Set is an array that will store the values of the Mandelbrot set calculated.
N = 1000
x = np.linspace(-2,2,N)
y = np.linspace(-2,2,N)
Complex_Values = np.empty([N,N],complex)
Mandelbrot_Set = np.empty([N,N], float)

#Define the Mandelbrot function
#I use an if statement if the absolute value of x is greater than 2 it will return 1.0.
#This is the max number of iterations it could obtain before returning a value greater than 2.
def Mandelbrot(c):
    z_p = lambda z: z*z + c
    x = z_p(0 + 0j)
    if abs(x) > 2:
        return 1.0
    else:
        for i in range(1000):
            x = z_p(x)
            if abs(x) > 2:
                return 1.0
            else:
                i +=1
    return 0
#A for loop is used to run the number of iterations wanted until the array is complete.
for i in range(N):
    for k in range(N):
        Complex_Values[i,k] = x[k] + 1j*y[i]
        Mandelbrot_Set[i,k] = Mandelbrot(Complex_Values[i,k])
    
#Plot graph in black and white scale
plt.figure(figsize=(20,20))
plt.title("Mandelbrot Set")
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.gray()
plt.imshow(Mandelbrot_Set,extent=[-2,2,-2,2])
plt.show()
```
