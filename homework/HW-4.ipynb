```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from numpy import exp
```

```python
#Write a program to calcualte C_v for a given value of T for a sample of 1000 cm^3 of aluminum
```

```python
#Define function to calculate heat capacity
def heatCapacity(T):
    
#Define constants
    V = 0.001 #Volume in meters cubed
    p = 6.022e28 #Density in meters cubed
    k = 1.38e-23 #Joules per Kelvin
    thetaD = 428 #Debye temperature in Kelvin

#Define a lambda function
#This is the integrand of the integral in the specific heat formula
    f = lambda x:((x**4)*exp(x))/(exp(x)-1)**2

#Define variable to hold the two quantities created by scipy's integrate function
    area = integrate.quad(f,0,thetaD/T)

#Equation to calculate and store value for the specific heat
    Cv = 9*V*p*k*(T/thetaD)**3*area[0]

#Print the heat capacity
    print("The heat capacity at temperature", T, "K is: ", Cv)
```

```python
heatCapacity(300)
```

```python
#Write a program to plot Cv as a function of temperature from T=5K to T=500K
```

```python
#Define constants
V = 0.001 #Volume in meters cubed
p = 6.022e28 #Density in meters cubed
k = 1.38e-23 #Joules per Kelvin
thetaD = 428 #Debye temperature in Kelvin

#State the range of the temperature
T = np.linspace(5, 500,1000)

#Define a lambda function
#This is the integrand of the integral in the specific heat formula
f = lambda x:((x**4)*exp(x))/(exp(x)-1)**2

#Create array to hold different heat capacity values for different temperatures
Cv = []

#Populate the array with this oop over a specified temperature range
for i in range(len(T)):
    Cv.append((((T[i]/thetaD)**3)*9*V*p*k)*(integrate.quad(f,0,thetaD/T[i]))[0])

#Plot the figure
plt.figure(figsize=(10,10))
plt.title('Heat Capacity vs. Temperature')
plt.xlabel('Temperature (K)')
plt.ylabel(r'$C_\nu$ (J/K)')
plt.plot(T,Cv)

plt.show()
```
