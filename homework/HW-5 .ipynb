```python
#Import relevant libraries.
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
%matplotlib inline

#Declaring the size of our figure.
plt.figure(figsize=[20,20])

#Define the function.
f = lambda x, a, b: a * np.exp(-(x**2-a**2)/b)

#Define the parameters.
x_min, x_max = -5.0, 5.0
N = 200
order1 = 2
order2 = 4
order3 = 6
order4 = 8
order5 = 10
order6 = 15
order7 = 20
order8 = 25
order9 = 35
a, b = 2.0, 3.0

#np arrays populate noisy data.
x = np.linspace(x_min, x_max, N)
y = f(x, a, b) + np.random.rand(N) - 0.5

#Polyfit for parameters 
fit1 = np.polyfit(x, y, order1, full=True)
y_val1 = np.polyval(fit1[0], x)

fit2 = np.polyfit(x, y, order2, full=True)
y_val2 = np.polyval(fit2[0], x)

fit3 = np.polyfit(x, y, order3, full=True)
y_val3 = np.polyval(fit3[0], x)

fit4 = np.polyfit(x, y, order4, full=True)
y_val4 = np.polyval(fit4[0], x)

fit5 = np.polyfit(x, y, order5, full=True)
y_val5 = np.polyval(fit5[0], x)

fit6 = np.polyfit(x, y, order6, full=True)
y_val6 = np.polyval(fit6[0], x)

fit7 = np.polyfit(x, y, order7, full=True)
y_val7 = np.polyval(fit7[0], x)

fit8 = np.polyfit(x, y, order8, full=True)
y_val8 = np.polyval(fit8[0], x)

fit9 = np.polyfit(x, y, order9, full=True)
y_val9 = np.polyval(fit9[0], x)

para, extras = curve_fit(f, x, y)

#Plots individual lines based on the fit.
plt.plot(x, f(x, a, b), '--', label = 'Ideal Function')
plt.plot(x, y, 'o', label = 'Noisy Data')
plt.plot(x, y_val1, label = 'Polyfit1order'+str(order1))
plt.plot(x, y_val2, label = 'Polyfit2order'+str(order2))
plt.plot(x, y_val3, label = 'Polyfit3order'+str(order3))
plt.plot(x, y_val4, label = 'Polyfit4order'+str(order4))
plt.plot(x, y_val5, label = 'Polyfit5order'+str(order5))
plt.plot(x, y_val6, label = 'Polyfit6order'+str(order6))
plt.plot(x, y_val7, label = 'Polyfit7order'+str(order7))
plt.plot(x, y_val8, label = 'Polyfit8order'+str(order8))
plt.plot(x, y_val9, label = 'Polyfit9order'+str(order9))

#Dimensions and labels for plot
plt.xlabel('x', fontsize = 16)
plt.ylabel('$f(x)$', fontsize = 16)
plt.xlim([x_min, x_max])
plt.legend(fontsize = 14)
plt.title('Gaussian Function and Fits')
plt.show()

#Prints out fitted coefficients and residuals for each plot.
print('Fitted coefficients for fit1:' + str(fit1[0]) + '\n Residuals: ' + str(fit1[1]))
print('Fitted coefficients for fit2:' + str(fit2[0]) + '\n Residuals: ' + str(fit2[1]))
print('Fitted coefficients for fit3:' + str(fit3[0]) + '\n Residuals: ' + str(fit3[1]))
print('Fitted coefficients for fit4:' + str(fit4[0]) + '\n Residuals: ' + str(fit4[1]))
print('Fitted coefficients for fit5:' + str(fit5[0]) + '\n Residuals: ' + str(fit5[1]))
print('Fitted coefficients for fit6:' + str(fit6[0]) + '\n Residuals: ' + str(fit6[1]))
print('Fitted coefficients for fit7:' + str(fit7[0]) + '\n Residuals: ' + str(fit7[1]))
print('Fitted coefficients for fit8:' + str(fit8[0]) + '\n Residuals: ' + str(fit8[1]))
print('Fitted coefficients for fit9:' + str(fit9[0]) + '\n Residuals: ' + str(fit9[1]))
```
