```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt

#Define function to plot
def plot_function(start, end, f, parts, resultsx, resultsy):
    x = np.linspace(start,end,100)
    y = f(x)
    plt.plot(x, y, 'r')
    plt.xlim([start, end])
    
    y2 = np.array([0,0])
    for i in range(parts):
        x1 = resultsx[i:i+2]
        y1 = resultsy[i:i+2]
        plt.fill_between(x1, y1, y2, where=y1>=y2, facecolor='gray')
        linex, liney = [resultsx[i],resultsx[i]], [0,resultsy[i]]
        plt.plot(linex, liney, color='black', linewidth=1.0)
                  
                  
    plt.show()

#Define function that uses the Rectangle rule
def Rectangle(start, end, parts, function):
    """
    
    Rectangle sum rule
    
    """
    
#Define the x and y points
    deltax = (end - start) / parts
    resultsx = np.linspace(start, end, parts)
    resultsy = function(resultsx)
    
#To calculate area
    area = np.empty([parts], float)
    for i in range(parts):
        area[i]= resultsy[i] * deltax
        
#Prints out text and the sum of the area using the Rectangle rule
    print("The sum of the area using the Rectangle Rule is: ", sum(area))

#Define function that uses the Trapezoid rule
def Trapezoid(start, end, parts, function):
    """
    
    Trapezoid sum rule
    
    """
    deltax = (end - start) / parts
    resultsx = np.linspace(start, end, parts+1)
    resultsy = function(resultsx)
    
#To calculate area using Trapezoid rule
    area = np.empty([parts], float)
    for i in range(parts):
            area[i] = (resultsy[i]+resultsy[i+1]) * deltax/2
    print("The sum of the area using the Trapezoid Rule is: ", sum(area))    

#Define function for Simpson's rule
def Simpson(start, end, parts, function, plot =1):
    """
    
    Rectangle sum rule
    
    """
    deltax = (end - start) / parts
    resultsx = np.linspace(start, end, parts+1)
    resultsy = function(resultsx)
    
#To calculate area using Simpson's rule    
#I was late submitting this assignment because I couldn't figure out why I was getting a negative value for this part.
#I was getting -8.88 value... Turns out I wasn't adding "+1". Took me forever to finally find my error. 
    area = np.empty([parts+1], float)
    for i in range(parts+1):
        if i == 0 or i == parts:
            coef = 1/3
        elif i%2 == 1:
            coef = 4/3
        else:
            coef = 2/3
        print(i,coef ==4/3)          
        area[i] = coef*resultsy[i] * deltax
    print("The sum of the area using Simpson's Rule is: ", sum(area))
        
    if plot == 1:
        plot_function(start, end, f, parts, resultsx, resultsy)
```
