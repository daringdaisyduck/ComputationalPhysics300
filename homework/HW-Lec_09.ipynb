```python
#Import relevant libraries.
import matplotlib.pyplot as plt
import numpy as np
from numpy import fft
import urllib.request
%matplotlib inline

req = urllib.request.Request('http://www-personal.umich.edu/~mejn/cp/data/circular.txt')
response = urllib.request.urlopen(req)
the_page = response.read().decode()
#data = the_page

data = np.fromstring(the_page, dtype = float, sep = '\n').reshape([501,501])

image = plt.imshow(data)
```

```python
#2D Fourier Transform on data using fft2 from numpy.
f = np.fft.fft2(data)
plt.figure(figsize=(20,10))
plt.title('Fourier Transform of Image')
plt.xlabel('Frequency')
plt.ylabel('Intensity')
plt.plot(np.abs(f),'k--')
plt.show()
```

The graph above shows the frequency space of the initial data. 

```python
#2D Fourier Transform on data using fft2 from numpy zoomed in on the left end.
f = np.fft.fft2(data)
plt.figure(figsize=(20,10))
plt.title('Fourier Transform of Image')
plt.xlabel('Frequency')
plt.ylabel('Intensity')
plt.plot(np.abs(f),'k--')
plt.xlim(0,20)
plt.show()
```

The graph above is a zoomed in version of the previous graph looking closely near  0.

```python
#Compressed data.
data2 = np.fft.ifft2(f).real
plt.imshow(np.abs(data2))
plt.title('Compressed Data')
plt.show()
plt.imsave('ifftimage.png',data)
```

Saving the data as a PNG file as shown above, we can see that the data is compressed.
