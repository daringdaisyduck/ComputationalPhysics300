```python
#Import relevant libraries.
%matplotlib inline
import numpy as np
from numpy import linspace
import matplotlib.pyplot as plt
from IPython.display import Audio
from scipy.io import wavfile
import os

workdir = str(os.getcwd())

os.chdir('/Users/daisy/Desktop')

rate, audio = wavfile.read('crow.wav')

N = audio.shape[0]
time = np.arange(N)/rate
amplitude = audio

print('Audio length =',round(time[-1],6),'seconds')

plt.figure(figsize=(12,3))
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.plot(time, amplitude)
plt.show()
```

```python
from scipy import signal,misc
import numpy as np

freqs, times, Sx = signal.spectrogram(audio, fs=rate, window='hanning', nperseg= 2, noverlap= N-21500)

plt.figure(figsize=(15,5))
plt.pcolormesh(times,freqs/10,np.log10(Sx))
plt.xlabel('Time (s)')
plt.ylabel('Frequency (kHz)');
```
