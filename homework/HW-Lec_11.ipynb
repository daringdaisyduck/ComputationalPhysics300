2) Error Analysis
Try to do the intergral many times (100) for different $N$ values (100, 1000, 10000, 100000), collect the data and calculate its standard varaince.

```python
#Import relevant libraries.
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from math import sin
from random import random

#Define the functions of the circles.
F1 = lambda x: np.sqrt(1 - (x-1)**2)
F2 = lambda x: 2 - np.sqrt(4 - x**2)

x = np.linspace (0, 2, 100)

#Define the Monte Carlo function.
def Monte_Carlo(N, F1, F2):
    count = 0
    for i in range(N):
        x = 2 * random()
        y = random()
        if y < F1(x) and y > F2(x):
            count += 1
    j = 2 * count/N
    k = count/N
    return(k, j)

#Plot the figure.
plt.figure(figsize = (10, 10))
plt.plot(x, F1(x), 'k')
plt.plot(x, F2(x), 'k')
plt.plot([0, 0, 2, 2, 0], [0, 1, 1, 0, 0], 'r')
plt.fill_between(x, F1(x), F2(x), where  = F1(x) > F2(x), color = 'green', alpha = 0.5)
plt.show()

#Print out calculated areas.
print ('Area in Shaded Region for N = 100', '=', str(Monte_Carlo(100, F1, F2)[1]))
print ('Area in Shaded Region for N = 1000', '=', str(Monte_Carlo(1000, F1, F2)[1]))
print ('Area in Shaded Region for N = 10000', '=', str(Monte_Carlo(10000, F1, F2)[1]))
print ('Area in Shaded Region for N = 100000', '=', str(Monte_Carlo(100000, F1, F2)[1]))
print ('Area in Shaded Region for N = 1000000', '=', str(Monte_Carlo(1000000, F1, F2)[1]))
```

```python
#Find the variance.
def variance(N):
    q = Monte_Carlo(N, F1, F2)[0]/N
    var = (q * (1 - q)/N)
    return var

#Print out the calculated variance for each value of N.
print ('The standard variance for N = 100', '=', variance(100))
print ('The standard variance for N = 1000', '=', variance(1000))
print ('The standard variance for N = 10000', '=', variance(10000))
print ('The standard variance for N = 100000', '=', variance(100000))
print ('The standard variance for N = 1000000', '=', variance(1000000))
```

As N increases we see that the accuracy of the area increases.
